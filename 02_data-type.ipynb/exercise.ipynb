- 8진수 : 0o
- 16진수 : 0x

```python
a = 100
print(a)

b = 0b11
print(b)   # 3

c = 0o23
print(c)   # 19

d = 0xff
print(d)   # 255
```

```python
#지수 표현식
a = 4.24e10
print(a)   # 42400000000.0
print(type(a))   # <class 'float'>

b = 4.24e-10
print(b)   # 4.24e-10
print(type(b))   # <class 'float'>
```

```python
origin = 'ohgiraffers'
with_white_space = '   oh giraffers  '

# 인자 생략 == 공백 제거
print(with_white_space.strip())    # oh giraffers

# '   o'까지 제거
print(with_white_space.strip('   o'))    # h giraffers

# 'os' 제거
print(origin.strip('os'))    # hgiraffer

# lstrip() - 선행만 제거
print(origin.lstrip('os'))    # hgiraffers

# rstrip() - 후행만 제거
print(origin.rstrip('os'))    # ohgiraffer
```
