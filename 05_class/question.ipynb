```python
"""### **문제 4: 네임스페이스 테스트 - 전역과 로컬**

카피바라는 이직 준비 도구를 사용하며, 전역 변수와 로컬 변수를 관리합니다. `namespace_test` 함수를 작성해 네임스페이스를 테스트하세요.

1. 전역 변수 `tool`을 "연필"로 설정하세요.
2.  함수 `namespace_test` 내부에 로컬 변수 `local_tool`을 "키보드"로 설정하고, `global` 키워드를 사용해 전역 변수를 "펜"으로 수정하세요.
3. 함수 내부에서 로컬 변수와 전역 변수의 값을 출력하세요."""

tool = "연필"


def namespace_test():
    local_tool = "키보드"

    global tool
    tool = "펜"

    print(f"로컬 변수 local_tool은 {local_tool}입니다.")
    print(f"전역 변수 tool은 {tool}입니다.")


namespace_test()
print(f"함수 실행 후 전역 변수 tool은 {tool}입니다.")
```

```python
"""### **문제 6: 네임스페이스 테스트 - nonlocal**

카피바라는 이직 준비 중 추가적인 정보가 필요합니다. 중첩 함수와 `nonlocal` 키워드를 사용해 바깥 함수의 변수를 수정하세요.

1. 함수 `nested_function`을 작성하세요.

1. 바깥 함수의 변수 `status`를 "작성 중"으로 초기화하세요.
2. 중첩 함수에서 `status`를 "완료"로 수정하세요."""
def nested_function():
    status = "작성 중"

    def update_status():
        nonlocal status
        status = "완료"

    update_status()
    print(status)


nested_function()
```

```python
"""
### 문제 1: **기본 상속을 활용한 클래스 생성**

카피바라는 이직을 준비하며 새로운 직업군을 탐색하고 있습니다. 카피바라의 기본적인 정보를 담는 `Person` 클래스를 만들고, 이를 상속받는 `Candidate` 클래스를 생성하세요.

1. `Person` 클래스:
    - 클래스 속성 `national`은 "Korea"로 설정합니다.
    - 메서드 `introduce`는 `"안녕하세요, 저는 한국에서 왔습니다."`를 반환합니다.
2. `Candidate` 클래스:
    - `Person` 클래스를 상속받습니다.
    - 생성자에서 이름(`name`)과 관심 분야(`field`)를 매개변수로 받아 저장합니다.
    - 메서드 `introduce`를 오버라이드하여 `"안녕하세요, 저는 [name]입니다. [field] 분야에 관심이 있습니다."`를 반환합니다.
"""
class Person:
    nation = "Korea"
    def introduce(self):
        print("안녕하세요, 저는 한국에서 왔습니다.")
class Candidate(Person):
    def __init__(self,name,field):
        super().__init__()
        self.name = name
        self.field = field
    def introduce(self):
        print(f"안녕하세요, 저는 {self.name} 입니다. {self.field} 분야에 관심이 있습니다.")
```

```python
"""### 문제 2: **다중 상속을 활용한 클래스 생성**

카피바라는 이직을 준비하며 여러 강사님들의 도움을 받고 있습니다. 강사의 역할을 분리한 클래스를 만들어 상속 구조를 테스트하세요.

1. `Mentor` 클래스:
    - 메서드 `guide`는 `"이직 준비를 도와드립니다."`를 반환합니다.
2. `Lecturer` 클래스:
    - 메서드 `teach`는 `"이직에 필요한 기술을 가르쳐드립니다."`를 반환합니다.
3. `Coach` 클래스:
    - `Mentor`와 `Lecturer` 클래스를 다중 상속받습니다.
    - 메서드 `support`는 `"여러 방면으로 지원을 아끼지 않겠습니다."`를 반환합니다."""
class Mentor:
    def guide(self):
        print("이직 준비를 도와드립니다.")
class Lecturer:
    def teach(self):
        print("이직에 필요한 기술을 가르쳐드립니다.")
class Coach(Mentor,Lecturer):
    def support(self):
        print("여러 방면으로 지원을 아끼지 않겠습니다.")
```

```python
"""### 문제 3: **부모 클래스 속성 오버라이딩**

카피바라는 이직 준비를 위해 개인적인 특성을 강조하고 있습니다. 부모 클래스의 메서드를 오버라이딩하여 각기 다른 인스턴스를 만들어보세요.

1. `Person` 클래스:
    - 속성 `name`과 `national`을 생성자로 받아 저장합니다.
    - 메서드 `introduce`는 `"안녕하세요, 저는 [name]입니다. 저는 [national]에서 왔습니다."`를 반환합니다.
2. `Candidate` 클래스:
    - `Person` 클래스를 상속받습니다.
    - `introduce` 메서드를 오버라이드하여 `"안녕하세요, 제 이름은 [name]이고, 이직 준비 중입니다."`를 반환합니다."""

class Person:
    def __init__(self,name,national):
        self.name = name
        self.national = national
    def introduce(self):
        print( f"안녕하세요, 저는 {self.name}입니다. 저는 {self.national}에서 왔습니다.")
class Candidate(Person):
    def __init__(self,name):
        self.name = name
    def introduce(self):
        print( f"안녕하세요, 저는 {self.name}이고, 아직 준비 중입니다.")
```

```python
"""## 문제 4: **상속의 특성 통합 테스트**

카피바라는 강사님들의 지도를 받으며 다양한 분야에서 경험을 쌓고 있습니다. 상속과 메서드 오버라이딩, 다중 상속을 종합적으로 테스트하세요.

1. `Mentor` 클래스:
    - 속성 `field`를 생성자로 받아 저장합니다.
    - 메서드 `guide`는 `"저는 [field] 분야의 멘토입니다."`를 반환합니다.
2. `Lecturer` 클래스:
    - 속성 `experience`를 생성자로 받아 저장합니다.
    - 메서드 `teach`는 `"저는 [experience]년의 강의 경력을 가지고 있습니다."`를 반환합니다.
3. `Candidate` 클래스:
    - `Mentor`와 `Lecturer` 클래스를 다중 상속받습니다.
    - `guide` 메서드를 오버라이드하여 `"저는 [field] 분야의 강사이며, 이직 준비를 도와드립니다."`를 반환합니다."""

class Mentor:
    def __init__(self,field):
        self.field=field
    def guide(self):
        print(f"저는 {self.field} 분야의 멘토입니다.")
class Lecturer:
    def __init__(self,experience):
        self.experience = experience
    def teach(self):
        print(f"저는 {self.experience}년의 강의 경력을 가지고 있습니다")
class Candidate(Mentor,Lecturer):
    def __init__(self,field,experience):
        Mentor().__init__(self,field)
        Lecturer().__init__(self,experience)

    def support(self):
        print(f"저는 {self.field} 분야의 강사이며, 이직 준비를 도와드립니다.")
```
