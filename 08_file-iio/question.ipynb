```python
import csv

datas = [
    ["회사명","직무","결과"],
    ["회사A","데이터 분석가","합격"],
    ["회사B","소프트웨어 엔지니어","불합격"],
    ["회사C","제품 관리자","진행 중"]
]

with open("application.csv",'w',encoding='utf-8') as f:
    writer = csv.writer(f)
    for data in datas:
        writer.writerow(data)
with open("application.csv",'r',encoding='utf-8') as f:
    print(f.read())
```

```python
with open("system.log",'w',encoding='utf-8') as f:
    f.writelines(
        [
            "[INFO] 2024-01-01 12:00:00: 시스템 시작\n",
            "[WARN] 2024-01-01 12:05:00: 메모리 사용량 증가\n",
            "[ERROR] 2024-01-01 12:10:00: 프로그램 종료\n",
        ]
    )
with open("system.log",'r',encoding='utf-8') as f:
    for line in f.readlines():
        if ("ERROR" in line) or ('WARN' in line):
            print(line)
```

```python
import json
with open("favorite.json",'w',encoding='utf-8') as f:
    json.dump(
        {
            "name" : "카피바라",
            "favorite_foods": ["도토리","호두","밤"]
        },
        f,
        ensure_ascii=False,
        indent=4
    )
    

with open("favorite.json",'r',encoding='utf-8') as f:
    print(json.load(f)["favorite_foods"])
```

```python

html_string = """
<html>
    <head><title>카피바라의 이력서</title></head>
    <body>
        <h1>카피바라의 이력서</h1>
        <p>경력: 신입</p>
    </body>
</html>
"""

with open("resume.html", "w", encoding="utf-8") as f:
    f.write(html_string)


with open("resume.html", "r", encoding="utf-8") as f:
    content = f.read()

content = content.replace(
    "<p>경력: 신입</p>",
    "<p>경력: 백엔드 개발자 1년차</p>"
)

with open("resume.html", "w", encoding="utf-8") as f:
    f.write(content)

with open("resume.html", "r", encoding="utf-8") as f:
    print(f.read())
```
