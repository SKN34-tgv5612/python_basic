```python
"""
## **문제 1: 손님 응대하기**

**"카피바라는 카페에서 아르바이트를 시작했다."**

카피바라는 손님들에게 인사를 해야 합니다.

1. 손님 리스트 `customers`에 다음과 같은 손님들이 있습니다.
    - `'토끼', '곰', '여우', '호랑이'`
2. `for`문을 사용하여 **각 손님에게 다음과 같은 메시지를 출력하세요**:
    - **"안녕하세요, [손님 이름]님! 무엇을 도와드릴까요?"**
"""

customers = ['토끼','곰', '여우','호랑이']

for customer in customers:
    print(f"안녕하세요, {customer}님! 무엇을 도와드릴까요?")
```

```python
"""
## **문제 2: 메뉴 준비하기**

**"카피바라는 주문받은 음료를 준비해야 한다."**

카피바라는 손님들이 주문한 음료를 준비합니다.

1. 주문 리스트 `orders`에 다음 음료가 있습니다.
    - `'아메리카노', '카페라떼', '핫초코'`
2. `for`문을 사용하여 **각 음료를 준비하는 메시지를 출력하세요**:
    - **"[음료 이름]를 준비 중입니다..."**
"""

orders = ['아메리카노','카페라떼','핫초코']
for order in orders:
    print(f'{order}를 준비 중입니다...')
```

```python
"""
## **문제 3: 청소하기**

**"일이 끝난 후, 카피바라는 테이블을 청소해야 한다."**

카피바라는 카페의 테이블을 하나씩 청소합니다.

1. 테이블 리스트 `tables`에 숫자 `1`부터 `5`까지의 테이블 번호가 있습니다.
2. `for`문을 사용하여 **각 테이블을 청소하는 메시지를 출력하세요.**
    - **"[n번 테이블] 청소 완료!"**
"""
for table in range(5):
    print(f"[{table+1}번 테이블] 청소 완료!")
```

```python
"""## **문제 4: 시간 기록하기**

**"카피바라는 출근한 후 시간마다 일을 기록해야 한다."**

카피바라는 9시부터 12시까지 1시간마다 기록을 남깁니다.

1. `for-range`를 사용하여 9부터 12까지의 시간을 반복하세요.
2. 각 시간마다 **"현재 시간: [시간]시, 근무 기록합니다."**를 출력하세요."""

for time in range(9,13):
    print(f"현제 시간 : {time}시, 근무 기록합니다.")
```

```python
"""## **문제 5: 손님 수 세기**

**"손님이 계속 들어오고 있습니다."**

카피바라는 10명의 손님이 들어올 때마다 번호를 세야 합니다.

1. `for-range`를 사용하여 숫자 `1`부터 `10`까지 출력하세요.
2. 각 손님마다 **"손님 [번호]번째 입장!"**을 출력하세요."""

for n in range(1,11):
    print(f"손님 {n}번째 입장!")
```

```python
"""## **문제 6: 커피 만들기**

**"주문받은 커피를 몇 잔 만들까?"**

카피바라는 주문받은 커피의 수량을 만들려고 합니다.

1. `for-range`를 사용해 숫자 `1`부터 `5`까지의 커피 번호를 반복하세요.
2. 각 커피마다 **"[커피 번호]번째 커피를 만들고 있습니다."**를 출력하세요."""

for n in range(1,6):
    print(f"{n}번째 커피를 만들고 있습니다.")
```

```python
"""## **문제 7: 테이블 청소하기**

**"카피바라는 닫기 전에 테이블을 모두 청소해야 한다."**

카피바라는 5번 테이블까지 순서대로 청소해야 합니다.

1. `while`문을 사용하여 `1`번부터 `5`번 테이블을 청소하세요.
2. 각 테이블마다 **"[n번 테이블] 청소 완료!"**를 출력하세요.
"""
n = 1
while n <= 5:
    print(f"{n}번 테이블 청소 완료!")
    n += 1
```

```python
"""## **문제 8: 손님이 그만 올 때까지 기다리기**

**"손님이 그만 올 때까지 카운터에서 기다리는 카피바라."**

카피바라는 손님이 그만 올 때까지 카운터를 지켜야 합니다.

1. 변수 `customers`에 초기값 `5`를 설정하세요.
2. `while`문을 사용해 손님이 줄어들 때마다 `customers`를 1씩 감소시키세요.
3. 손님이 0이 될 때까지 **"남은 손님 수: [customers]명"**을 출력하세요.
4. 손님이 모두 떠나면 **"오늘의 영업이 끝났습니다!"**를 출력하세요."""
customers = 5
while customers > 0:
    print(f"남은 손님 수: {customers}명 ")
    customers -= 1
print("오늘의 영업이 끝났습니다!")
```

```python
"""## **문제 9: 알바비 계산하기**

**"카피바라는 일한 시간만큼 알바비를 계산합니다."**

카피바라의 시급은 **1시간에 1000원**입니다.

1. 변수 `work_hours`에 일한 시간(정수)을 입력받으세요.
2. `while`문을 사용해 일한 시간만큼 반복하며 `pay`에 시급을 더하세요.
3. 최종 알바비를 **"오늘의 알바비는 [pay]원입니다."**로 출력하세요."""

work_hours = int(input())
pay = 0
while work_hours > 0:
    work_hours -= 1
    pay += 1000 
print(f'오늘의 알바비는 {pay}원입니다.')
```

```python
"""## **문제 1: 손님 응대 중단하기**

**"카피바라는 갑자기 피곤해져서 손님 응대를 중단하고 싶다."**

카피바라는 카페에서 손님을 맞이하고 있습니다. 하지만 카피바라는 너무 피곤해서 **3번째 손님**까지만 응대하고 싶습니다.

1. 손님 리스트 `customers`에 `'토끼', '곰', '여우', '호랑이', '늑대'`를 저장하세요.
2. `for`문을 사용해 각 손님을 응대하며 다음 메시지를 출력하세요:
    - **"안녕하세요, [손님 이름]님! 무엇을 도와드릴까요?"**
3. 3번째 손님까지 응대한 후, 반복문을 종료하고 **"죄송합니다, 오늘은 여기까지입니다!"**를 출력하세요.
"""

customers = ['토끼', '곰', '여우', '호랑이', '늑대']
for idx , customer in enumerate(customers):
    print(f'안녕하세요, {customer}님! 무엇을 도와드릴까요?')
    if idx == 2:
        print('죄송합니다, 오늘은 여기까지입니다!')
        break
    
```

```python
"""
## **문제 2: 주문 실수 중단하기**

**"카피바라는 주문을 받다가 실수를 발견하면 즉시 중단하려고 한다."**

카피바라는 손님들의 주문을 확인하고 있습니다. 주문에 문제가 있으면 즉시 중단해야 합니다.

1. 주문 리스트 `orders`에 `'아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소'`를 저장하세요.
2. `for`문을 사용해 각 주문을 확인하며 다음과 같이 출력하세요.
    - **"[음료 이름] 주문 확인했습니다!"**
3. 만약 주문이 `'모르겠어요'`이면, **"주문에 문제가 있습니다! 확인이 필요합니다."**를 출력하고 반복문을 종료하세요."""

orders = ['아메리카노', '카페라떼', '핫초코', '모르겠어요', '에스프레소']

for order in orders:
    print(f'[{order}] 주문 확인했습니다!')
    if order == "모르겠어요":
        print(f"주문에 문제가 있습니다! 확인이 필요합니다.")
        break 
```

```python
"""
## **문제 3: 홀수 테이블 청소 건너뛰기**

**"카피바라는 짝수 테이블만 청소하고 싶다."**

카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.

카피바라는 홀수 테이블은 건너뛰고 짝수 테이블만 청소하려고 합니다.

1. 테이블 리스트는 `1`부터 `6`까지의 숫자로 구성되어 있습니다.
2. `for`문을 사용해 각 테이블 번호를 확인하세요.
3. 테이블 번호가 **홀수**이면 해당 테이블을 건너뛰세요.
4. 짝수 테이블인 경우 **"[n번 테이블] 청소 완료!"**를 출력하세요.
"""
for table_number in range(1,7):
    if table_number % 2 == 0:
        print(f"{table_number}번 테이블 청소 완료!")
```
