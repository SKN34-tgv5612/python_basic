```python
try:
    (실행할 코드)
except:
    (모든 예외 발생 시 동작할 코드)

```

````python
try:
    4 / 0
except:
    print("예외 발생")

# 실행 결과: 예외 발생
````

````python
try:
    a = [1,2]
    print(a[3])
    4/0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")
except IndexError:
    print("인덱싱 할 수 없습니다.")

# 실행 결과: 인덱싱 할 수 없습니다.
````

````python
try:
    4 / 0
except ZeroDivisionError as e:
    print(e)

# 실행 결과: division by zero
````

````python
f = None

try:
    f = open('foo.txt', 'r')
    # 동작할 코드 내용 작성

finally:
    if f:
        f.close()   # try 블럭 수행 중간에 오류가 발생하더라도 마지막에 무조건 실행된다.
````

```python
try:
    (동작할 코드)
except [발생오류 [as 오류변수]]:
    (오류발생 시 동작할 코드)
else:
    (오류가 발생하지 않으면 동작할 코드)
```

````python
try:
    age=int(input('나이를 입력하세요: '))
except:
    print('입력이 정확하지 않습니다.')
else:
    if age <= 18:
        print('미성년자는 출입금지입니다.')
    else:
        print('환영합니다.')
````
