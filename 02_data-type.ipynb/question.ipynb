```python
"""
### 문제 3

다음 조건에 맞게 변수를 선언하고, 주어진 논리 조건식의 결과를 출력하세요.

1. `name` 에 문자열 `'카피바라'`를 저장하세요.
2. `age` 에 정수 `20`을 저장하세요.
3. `gender` 에 문자열 `'male'`을 저장하세요.
4. 아래 조건을 각각 검사하여 `print()`로 결과를 출력하세요.
    - 이름이 `'카피바라'`이면 `True`
    - 나이가 **19세 이상이면 성인**으로 판단하고, 성인이면 `True`
    - 성인이면서 남성이면 `True`
    - 이름이 `'호랑이'`이고, 성인이며, **남성**이면 `True`
    - 이름이 `'카피바라'`이거나 성인이면 `True`
    - 이름이 `'카피바라'`가 아니라면 `True` (`not` 사용)
"""

name = '카피바라'
age = 20
gender = 'male'

print(name == '카피바라')  # True
print(age >= 19)          # True
print(age >= 19 and gender == 'male')  # True
print(name == '호랑이' and age >= 19 and gender == 'male')  # False
print(name == '카피바라' or age >= 19)  # True
print(not name == '카피바라')  # False
```

```python
javaTeam = {'Squirrel','Gorilla', 'Tiger', 'Sheep', 'monkey', 'wolf'}
pythonTeam = {'Pig', 'Squirrel', 'Bear', 'Gorilla'}

ohgiraffers = javaTeam.union(pythonTeam)
print(javaTeam | pythonTeam)    # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel', 'Pig', 'Bear'}
print(ohgiraffers)              # {'wolf', 'Sheep', 'Tiger', 'Gorilla', 'monkey', 'Squirrel', 'Pig', 'Bear'}
```

```python
print(javaTeam & pythonTeam)                # {'Squirrel', 'Gorilla'}
print(javaTeam.intersection(pythonTeam))    # {'Squirrel', 'Gorilla'}
```

```python
print(javaTeam - pythonTeam)              # {'monkey', 'wolf', 'Sheep', 'Tiger'}
print(javaTeam.difference(pythonTeam))    # {'monkey', 'wolf', 'Sheep', 'Tiger'}
```

```python
print(javaTeam ^ pythonTeam)                        # {'wolf', 'Sheep', 'Tiger', 'monkey', 'Pig', 'Bear'}
print(javaTeam.symmetric_difference(pythonTeam))    # {'wolf', 'Sheep', 'Tiger', 'monkey', 'Pig', 'Bear'}
```
