```python
print("=== OhGiraffers vending machine ===")
print("tiger: 우리는 1인 24시간 교대제로 간다")
print("monkey: 24시간 문의 환영, 다른 말로 하면?")
print("squirrel: 저도 하루에 너다섯 시간 자는데요?")
print("capybara: 살아남는 사람이 강한 사람입니다.")
print("====================================")
print("만나고 싶은 선생님을 입력해 주세요")
input_teacher = input()
print("====================================")

sleep_time = 0
match input_teacher:
    case "tiger":
        print("호랑이 선생님의 메타버스에 오신 것을 환영합니다.")
        sleep_time = 3
    case "monkey":
        print("원숭이 선생님의 밀림에 오신 것을 환영합니다.")
        sleep_time = 4
    case "squirrel":
        print("다람쥐 선생님의 매력의 늪에 오신 것을 환영합니다.")
        sleep_time = 5
    case "capybara" :  
        print("카피바라 선생님의 교실에 오신 것을 환영합니다.")
        sleep_time = 6
    case _:
        print("공부할 마음이 없으신 건 아니죠?")
print("매일 " + str(sleep_time) + "시간의 수면이 예상됩니다.")
# print("매일", sleep_time, "시간의 수면이 예상됩니다.")
```

```python
"""## **문제 1: 선택의 시작**

**"카피바라는 서울에 갈 것인가?"**

카피바라는 서울로 상경할 계획을 세우고 있습니다. 다음 조건을 확인하세요.

1. 사용자에게 "서울에 갈 계획이 있나요? (y/n): "를 입력받으세요.
2. 입력값이 `'y'`이면 **"서울로 간다!"**를 출력하세요.
"""
n = input("서울에 갈 계획이 있나요? (y/n): ")
print ("서울로 간다!" if n == "y" else "서울로 안간다!")
```

```python
"""## **문제 2: 짐 싸기**

**"카피바라는 짐을 싸야 할까?"**

카피바라는 서울에 가기로 마음먹었지만 짐을 싸야 할지 고민하고 있습니다.

1. 사용자에게 "집에 있는 가구를 다 가져갈까요? (y/n): "를 입력받으세요. 
2. 입력값이 `'n'`이면 **"서울에서 가구를 새로 산다."**를 출력하세요.

"""
n = input("집에 있는 가구를 다 가져갈까요? (y/n): ")
print ("서울에서 가구를 새로 산다." if n == "n" else "서울에서 가구를 새로 안 산다.")
```

```python
"""## **문제 3: 기차표 구매**

**"카피바라는 기차표를 살 수 있을까?"**

카피바라는 서울행 기차표를 구매하려 합니다.

1. 사용자에게 "현재 소지금은 얼마인가요? (숫자 입력): "을 입력받으세요.
2. 입력값이 `50000` 이상이면 **"기차표를 구매했다!"**를 출력하세요.

"""
n = int(input("현재 소지금은 얼마인가요? (숫자 입력): "))
print ("차표를 구매했다!" if n >= 50000 else "차표를 구매 못했다!")
```

```python
"""## **문제 4: 서울 도착**

**"카피바라는 서울에 도착했다. 날씨가 좋을까?"**

서울에 도착한 카피바라는 날씨에 따라 이동할 계획입니다.

1. 사용자에게 "오늘 서울 날씨는 맑음인가요? (y/n): "를 입력받으세요.
2. 입력값이 `'y'`이면 **"서울 구경을 나간다."**를 출력하세요.
3. 입력값이 `'n'`이면 **"집에서 쉰다."**를 출력하세요.

"""
n = int(input("오늘 서울 날씨는 맑음인가요? (y/n)"))
print ("서울 구경을 나간다." if n >= "y" else "집에서 쉰다.")
```

```python
"""
## **문제 5: 집 구하기**

**"카피바라는 어디에 살까?"**

서울에서 집을 구해야 하는 카피바라는 예산에 따라 집을 선택합니다.

1. 사용자에게 "월세 예산은 얼마인가요? (숫자 입력): "를 입력받으세요.
2. 입력값이 `1000000` 이하이면 **"고시원에서 산다."**를 출력하세요.
3. 그 외의 경우 **"원룸을 계약한다."**를 출력하세요.

"""

n = int(input("월세 예산은 얼마인가요? (숫자 입력): "))
print ("고시원에서 산다." if n <= 1000000 else "원룸을 계약한다.")
```

```python
"""
## **문제 6: 직장 찾기**

**"카피바라는 직장을 구해야 한다!"**

카피바라는 서울에서 일자리를 찾고 있습니다. 직장에 따라 일의 강도가 다릅니다.

1. 사용자에게 "원하는 직장 종류를 입력하세요 (프리랜서/중소기업/대기업): "를 입력받으세요.
2. 입력값이 `'프리랜서'`이면 **"시간은 자유롭지만 불안정하다."**를 출력하세요.
3. 입력값이 `'중소기업'`이면 **"안정적이지만 일이 많다."**를 출력하세요.
4. 그 밖에 입력값에 대해서는 `“좋은 직장을 가지고 계시는군요”`를 출력하세요.

"""
n = int(input("원하는 직장 종류를 입력하세요 (프리랜서/중소기업/대기업): "))
match n:
    case "프리랜서":
        print("시간은 자유롭지만 불안정하다.")
    case "중소기업":
        print("안정적이지만 일이 많다.")
    case _:
        print("좋은 직장을 가지고 계시는군요")
```

```python
"""
## **문제 7: 생활 루틴 정하기**

**"카피바라는 서울에서 생활 루틴을 정해야 한다."**

서울 생활에 적응한 카피바라는 하루 루틴을 정하려 합니다.

1. 사용자에게 "아침에 일어나는 시간을 입력하세요 (숫자 입력, 6~10): "를 입력받으세요.
2. 입력값이 6이면 **"운동을 한다."**를 출력하세요.
3. 입력값이 7이면 **"공부를 한다."**를 출력하세요.
4. 그 외에는 **"바로 출근 준비를 한다."**를 출력하세요.

"""
n = int(input("아침에 일어나는 시간을 입력하세요 (숫자 입력, 6~10): "))
match n:
    case 6:
        print("운동을 한다.")
    case 7:
        print("공부를 한다.")
    case 8 | 9 | 10:
        print("바로 출근 준비를 한다.")
```

```python
"""
## **문제 8: 주말 계획**

**"카피바라의 주말 계획은?"**

주말이 되자 카피바라는 여유로운 시간을 보내려 합니다.

1. 사용자에게 "오늘 날씨가 어떤가요? (맑음/비/눈): "를 입력받으세요.
2. 입력값이 `'맑음'`이면 **"한강에 가서 피크닉을 즐기는 척하며 일을 한다."**를 출력하세요.
3. 입력값이 `'비'`이면 **"집에서 빗소리를 ASMR 삼아 일을 한다."**를 출력하세요.
4. 입력값이 `'눈'`이면 **"전기장판 안에서 귤을 까먹으며 일을 한다."**를 출력하세요.
5. 그 외의 값이 입력되면 **"암막 커튼 때문에 일어나지 못했다. 푹 잘 수 있군, 럭키비키잖아?"**를 출력하세요.
"""
n = int(input("오늘 날씨가 어떤가요? (맑음/비/눈): "))
match n:
    case "맑음":
        print("한강에 가서 피크닉을 즐기는 척하며 일을 한다.")
    case "비":
        print("집에서 빗소리를 ASMR 삼아 일을 한다.")
    case "눈":
        print("전기장판 안에서 귤을 까먹으며 일을 한다.")
    case _:
        print("암막 커튼 때문에 일어나지 못했다. 푹 잘 수 있군, 럭키비키잖아?")
```
